```python
credits_rc = credits.rename(index= str, columns= {"movie_id" : "id"})
movies_cm = movie_df.merge(credits_rc, on = "id")
movies_cm.head()
```

```python
movies_cln = movies_cm.drop(columns=["homepage", "title_x", "title_y", "status", "production_countries"])
movies_cln.head()
```

<b>Weighted Rating (Average Rating)</b>

W = Rv + Cm / v + m 

W = weighted rating <br>
R = penilaian dr range 0 to 10 (mean) -> rating <br>
v = votes <br>
C = nilai tengah yang bersebrangan <br>
m = minimum votes <br>


```python
v = movies_cln['vote_count']
R = movies_cln['vote_average']
C = movies_cln['vote_average'].mean()
m = movies_cln['vote_count'].quantile(0.70)
```

```python
movies_cln['weighted_average'] = ((R * v) + (C * m)) / (v +m)
```

```python
movies_cln.head()
```

```python
movie_sorted_ranking = movies_cln.sort_values('weighted_average', ascending= False)
movie_sorted_ranking[['original_title', 'vote_count', 'vote_average', 'weighted_average', 'popularity']].head(100)
```

```python
# plotting weighted average "sorting movie rank asc = false"

import matplotlib.pyplot as plt
import seaborn as sns

weighted_average = movie_sorted_ranking.sort_values('weighted_average', ascending= False)
plt.figure(figsize = (12, 6))
sns.barplot(x = weighted_average['weighted_average'].head(10), y = weighted_average['original_title'].head(10), data = weighted_average)
plt.xlim(5, 10)
plt.title('Best Movies by average votes')
plt.xlabel('Weighted Average Score')
plt.ylabel('Movie Title')
plt.savefig('best_movies.png')
```

```python
popularity = movie_sorted_ranking.sort_values('popularity', ascending = False)
plt.figure(figsize = (12, 6))
sns.barplot(x = popularity['popularity'].head(10), y = popularity['original_title'].head(10), data= popularity)

plt.title("Most Popular by Votes")
plt.xlabel("Score of Popularity")
plt.ylabel("Movie Title")
plt.savefig('best_popular_movies.png')
```
